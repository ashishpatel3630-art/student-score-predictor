```python
result = df.groupby("Hours_Studied")["Final_Score"].mean().reset_index()
plt.figure(figsize=(8,5))

plt.bar(result["Hours_Studied"], result["Final_Score"])

plt.title("Average Final Score vs Hours Studied")
plt.xlabel("Hours Studied")
plt.ylabel("Average Final Score")

plt.show()
```
